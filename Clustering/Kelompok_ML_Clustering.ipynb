```python
import re
import unicodedata
import numpy as np
import pandas as pd

def _to_float_or_none(s):
    try:
        return float(s)
    except:
        return None

def _norm_hemisphere(val, hemi):
    # E positive, W negative
    if hemi == 'W':
        return -abs(val)
    if hemi == 'E':
        return abs(val)
    return val

def _wrap_lon(val):
    # Map ke [-180, 180] tanpa NaN
    return ((val + 180.0) % 360.0) - 180.0

def parse_longitude_aggressive(x):
    """
    Kembalikan (float_value, parse_flag)
    parse_flag: 'direct' | 'eu' | 'us' | 'multi-dot' | 'dms' | 'digits-fallback' | 'fallback'
    """
    if pd.isna(x):
        # Akan ditangani di fallback, tetap kembalikan None
        return None, 'fallback'
    s_orig = str(x)

    # Normalisasi unicode & spasi
    s = unicodedata.normalize('NFKC', s_orig).strip()
    s = s.replace('\xa0', '').replace(' ', '')
    s = s.replace('\u2212', '-')  # unicode minus → ASCII

    # Ambil & buang hemisphere suffix
    hemi = None
    m_hemi = re.search(r'([NSEW])$', s, flags=re.IGNORECASE)
    if m_hemi:
        hemi = m_hemi.group(1).upper()
        s = re.sub(r'([NSEW])$', '', s, flags=re.IGNORECASE)

    # 1) Direct float
    val = _to_float_or_none(s)
    if val is not None:
        return _wrap_lon(_norm_hemisphere(val, hemi)), 'direct'

    # 2) DMS ringkas (ada derajat/menit/detik)
    if any(sym in s_orig for sym in ['°', "'", '′', '"', '″']):
        # Ekstrak D, M, S
        # Contoh cocok: 123°45'56.7", 123°45', 123° (long: umumnya 0-180)
        s_d = re.sub(r'[^0-9\.\,\-\+]', ' ', s_orig)  # sisakan angka, +-, ., ,
        parts = [p for p in re.split(r'\s+', s_d) if p]
        # ambil hingga 3 komponen angka
        nums = []
        for p in parts:
            # normalisasi koma ke titik
            p2 = p.replace(',', '.')
            try:
                nums.append(float(p2))
            except:
                pass
            if len(nums) == 3:
                break
        if len(nums) >= 1:
            deg = nums[0]
            minu = nums[1] if len(nums) >= 2 else 0.0
            sec  = nums[2] if len(nums) >= 3 else 0.0
            val = abs(deg) + minu/60.0 + sec/3600.0
            # tanda derajat mengikuti tanda deg
            val = -val if deg < 0 else val
            val = _norm_hemisphere(val, hemi)
            return _wrap_lon(val), 'dms'

    # 3) Format EU/US (punya koma/ titik)
    if ',' in s and '.' in s:
        last_dot = s.rfind('.')
        last_com = s.rfind(',')
        if last_com > last_dot:
            # EU: 1.234,56
            test = s.replace('.', '').replace(',', '.')
        else:
            # US: 1,234.56
            test = s.replace(',', '')
        val = _to_float_or_none(test)
        if val is not None:
            val = _norm_hemisphere(val, hemi)
            return _wrap_lon(val), 'eu' if last_com > last_dot else 'us'

    # 4) Hanya koma → desimal
    if ',' in s and '.' not in s:
        test = s.replace(',', '.')
        val = _to_float_or_none(test)
        if val is not None:
            val = _norm_hemisphere(val, hemi)
            return _wrap_lon(val), 'eu'

    # 5) Pola multi-dot '1.297.205' (hapus titik & sisipkan desimal setelah 3 digit pertama)
    if s.count('.') >= 2 and ',' not in s and re.fullmatch(r'[+\-]?[0-9.]+', s):
        sign = ''
        if s.startswith(('+', '-')):
            sign, s = s[0], s[1:]
        digits = s.replace('.', '')
        if digits.isdigit() and len(digits) >= 4:
            test = f"{sign}{digits[:3]}.{digits[3:]}"
            val = _to_float_or_none(test)
            if val is not None:
                val = _norm_hemisphere(val, hemi)
                return _wrap_lon(val), 'multi-dot'

    # 6) Ambil potongan numeric paling informatif dari string (last-resort)
    #    - Ambil semua digit (dengan opsional 1 tanda +/- di depan)
    m_num = re.findall(r'[+\-]?\d+', s)
    if m_num:
        sign = ''
        if s.startswith('-'):
            sign = '-'
        digits = ''.join([d.lstrip('+-') for d in m_num])
        # Jika sangat panjang, asumsikan derajat.xxx → sisipkan desimal setelah 3 digit
        if len(digits) >= 4:
            test = f"{sign}{digits[:3]}.{digits[3:]}"
        else:
            # <=3 digit → anggap integer derajat (mis. 107)
            test = f"{sign}{digits}"
        val = _to_float_or_none(test)
        if val is not None:
            val = _norm_hemisphere(val, hemi)
            return _wrap_lon(val), 'digits-fallback'

    # 7) Fallback terakhir: jika benar-benar tidak ada angka, kembalikan 0.0 (agar 0 NaN)
    return 0.0, 'fallback'

# ===== Terapkan ke kolom longitude =====
assert 'longitude' in df.columns, "Kolom 'longitude' tidak ditemukan."

df['longitude_raw'] = df['longitude']
parsed_vals = []
parse_flags = []

for v in df['longitude_raw'].tolist():
    val, flag = parse_longitude_aggressive(v)
    parsed_vals.append(val)
    parse_flags.append(flag)

df['longitude'] = parsed_vals
df['longitude_parse_flag'] = parse_flags

# Pastikan 0 NaN
nan_ct = pd.isna(df['longitude']).sum()
if nan_ct > 0:
    # Guarantee no NaN by filling with 0.0 (shouldn't happen with logic above)
    df['longitude'] = df['longitude'].fillna(0.0)

print("Konversi longitude selesai ✅")
print("Ringkasan parse_flag:")
print(pd.Series(df['longitude_parse_flag']).value_counts())

# Cek sebaran & contoh
print("Min/Max longitude:", float(np.min(df['longitude'])), float(np.max(df['longitude'])))
print(df[['longitude_raw','longitude','longitude_parse_flag']].head(10))
```

# K-Means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Mapping nama kolom di data ke nama standar
colmap = {
    'mag':       'mag',
    'depth':     'depth',
    'longitude': 'longitude',
    'latitude':  'latitude'
}
for k,v in colmap.items():
    if v not in df.columns:
        raise ValueError(f"Kolom '{v}' (untuk {k}) tidak ditemukan di df")

# Ambil 4 fitur inti
D = df[[colmap['mag'], colmap['depth'], colmap['longitude'], colmap['latitude']]].copy()
D = D.dropna().reset_index(drop=True)
D.columns = ['mag','depth','longitude','latitude']

# DETEKSI: Z-SCORE
means = D[['mag','depth','longitude','latitude']].mean()
stds  = D[['mag','depth','longitude','latitude']].std(ddof=0)
is_already_std = np.all(np.abs(means) < 0.15) and np.all((stds > 0.85) & (stds < 1.15))

# Parameter skala (μ,σ) bila tersedia
mu_hint = {}
sd_hint = {}

for nm in ['mag_mean','mag_std','depth_mean','depth_std']:
    if nm in df.columns:
        try:
            val = float(df[nm].iloc[0])
            if 'mag_' in nm:
                mu_hint['mag'] = val if 'mean' in nm else mu_hint.get('mag', None)
                sd_hint['mag'] = val if 'std'  in nm else sd_hint.get('mag', None)
            if 'depth_' in nm:
                mu_hint['depth'] = val if 'mean' in nm else mu_hint.get('depth', None)
                sd_hint['depth'] = val if 'std'  in nm else sd_hint.get('depth', None)
        except Exception:
            pass

scaler = StandardScaler()
if is_already_std:
    Xz = D[['mag','depth','longitude','latitude']].to_numpy()
    mu = {'mag':0.0,'depth':0.0,'longitude':0.0,'latitude':0.0}
    sd = {'mag':1.0,'depth':1.0,'longitude':1.0,'latitude':1.0}
else:
    Xz = scaler.fit_transform(D[['mag','depth','longitude','latitude']].to_numpy())
    mu = dict(zip(['mag','depth','longitude','latitude'], scaler.mean_))
    sd = dict(zip(['mag','depth','longitude','latitude'], scaler.scale_))

# K-Means Clustering
k = 3
km = KMeans(n_clusters=k, n_init=50, random_state=42)
labels = km.fit_predict(Xz)
D['cluster_id'] = labels
sil = silhouette_score(Xz, labels)
print(f"Silhouette score: {sil:.3f}")

centroids_z = km.cluster_centers_

# Skala
def label_by_threshold_original(mw, depth_km):
    if depth_km <= 60:
        return 'kecil' if mw < 4.5 else ('sedang' if mw < 6.0 else 'besar')
    elif depth_km <= 120:
        return 'kecil' if mw < 4.7 else ('sedang' if mw < 6.2 else 'besar')
    else:
        return 'kecil' if mw < 4.9 else ('sedang' if mw < 6.4 else 'besar')

# --- Cabang A: Kita tahu μ/σ asli (karena kita yang scaling) ---
if not is_already_std:
    centroids_orig = np.column_stack([
        centroids_z[:,0]*sd['mag']      + mu['mag'],
        centroids_z[:,1]*sd['depth']    + mu['depth'],
        centroids_z[:,2]*sd['longitude']+ mu['longitude'],
        centroids_z[:,3]*sd['latitude'] + mu['latitude'],
    ])
    # Map cluster → label via threshold skala asli (pakai (mag, depth) centroid)
    cluster_label_map = {
        i: label_by_threshold_original(mw=float(row[0]), depth_km=float(row[1]))
        for i,row in enumerate(centroids_orig)
    }
    D['cluster_label'] = D['cluster_id'].map(cluster_label_map)

# --- Cabang B: Data sudah z-score ---
else:
    # 2 opsi:
    # (B1) Jika μ/σ asli disediakan (mu_hint/sd_hint), konversi ambang ke z lalu label berdasar centroid_z
    if all(k in mu_hint for k in ['mag','depth']) and all(k in sd_hint for k in ['mag','depth']):
        to_z = lambda x, f: (x - mu_hint[f]) / sd_hint[f]
        thr = {
            "depth": {"shallow_max": to_z(60, "depth"), "mid_max": to_z(120, "depth")},
            "mag": {
                "kecil_sedang_shallow": to_z(4.5, "mag"), "sedang_besar_shallow": to_z(6.0, "mag"),
                "kecil_sedang_mid":     to_z(4.7, "mag"), "sedang_besar_mid":     to_z(6.2, "mag"),
                "kecil_sedang_deep":    to_z(4.9, "mag"), "sedang_besar_deep":    to_z(6.4, "mag"),
            }
        }
        def label_by_threshold_z(mz, dz):
            if dz <= thr["depth"]["shallow_max"]:
                return 'kecil' if mz < thr["mag"]["kecil_sedang_shallow"] else ('sedang' if mz < thr["mag"]["sedang_besar_shallow"] else 'besar')
            elif dz <= thr["depth"]["mid_max"]:
                return 'kecil' if mz < thr["mag"]["kecil_sedang_mid"] else ('sedang' if mz < thr["mag"]["sedang_besar_mid"] else 'besar')
            else:
                return 'kecil' if mz < thr["mag"]["kecil_sedang_deep"] else ('sedang' if mz < thr["mag"]["sedang_besar_deep"] else 'besar')
        cluster_label_map = {i: label_by_threshold_z(c[0], c[1]) for i,c in enumerate(centroids_z)}
        D['cluster_label'] = D['cluster_id'].map(cluster_label_map)

    # (B2) Jika μ/σ asli TIDAK tersedia: gunakan ranking fisik di ruang z
    else:
        # skor = (+mag_z) - (depth_z): makin besar magnitudo & makin dangkal → skor tinggi
        scores = centroids_z[:,0] - centroids_z[:,1]  # kolom 0=mag_z, 1=depth_z
        order = np.argsort(-scores)  # terbesar → "besar"
        rank_map = {order[0]:'besar', order[1]:'sedang', order[2]:'kecil'}
        D['cluster_label'] = D['cluster_id'].map(rank_map)

if not is_already_std:
    D['label_rule'] = [label_by_threshold_original(m, d) for m,d in zip(D['mag'], D['depth'])]

# 6) VISUALISASI: Scatterplot
label_to_num = {'kecil':0, 'sedang':1, 'besar':2}
cvals = [label_to_num.get(x, -1) for x in D['cluster_label']]

plt.figure(figsize=(7,6))
plt.scatter(D['longitude'], D['latitude'], c=cvals, s=14, alpha=0.9)
plt.xlabel('Longitude'); plt.ylabel('Latitude')
plt.title('K-Means pada [mag, depth, lon, lat] — warna: kecil/sedang/besar')

for name, num in label_to_num.items():
    plt.scatter([], [], label=name)
plt.legend(title='Kategori')
plt.show()

print("\nRingkasan label:")
print(D['cluster_label'].value_counts(dropna=False))
print("\nContoh 5 baris:")
print(D[['mag','depth','longitude','latitude','cluster_id','cluster_label']].head())

df_out = df.copy()
df_out = df_out.join(D[['cluster_id','cluster_label']], how='left')
```

# Alur Kerja Singkat

1. **Load data** ke `df` dan pastikan kolom inti ada: `mag`, `depth`, `longitude`, `latitude`.
2. **Backup koordinat asli**: simpan `longitude_orig`, `latitude_orig` sebelum ada proses standardisasi agar nilai derajat tidak hilang.
3. **Bentuk subset fitur** `D = [mag, depth, longitude, latitude]`, simpan **index baris asli** (`idx_keep`) sebelum `reset_index()` supaya baris yang diplot tetap sejajar dengan data yang diklaster.
4. **Standardisasi adaptif** (hanya 4 fitur):

   * Jika mean≈0 & std≈1 → data sudah z-score → pakai apa adanya.
   * Jika belum → `StandardScaler.fit_transform()` → simpan `μ` & `σ` per fitur untuk inverse ke unit asli.
5. **K-Means (k=3)** pada data terstandardisasi → dapat `cluster_id` dan nilai **silhouette**.
6. **Pelabelan manusiawi** `{kecil, sedang, besar}`:

   * **Prefer**: inverse centroid ke **unit asli** (pakai `μ, σ`) lalu bandingkan dengan **ambang 2-variabel** (lihat di bawah).
   * **Fallback** (jika μ,σ tidak ada): ranking fisik di z-space → klaster dengan **magnitudo lebih besar** & **kedalaman lebih dangkal** diberi label **“besar”**, berikutnya **“sedang”**, lalu **“kecil”**.
7. **Visualisasi**: scatter **longitude vs latitude** memakai **nilai asli** (prioritas `*_orig`; jika tidak ada → inverse dari z). Warna titik = `cluster_label`.
8. **Simpan kembali** `cluster_id` & `cluster_label` ke `df` berdasarkan `idx_keep`.

# Rasional Ambang (Magnitude × Depth)

* **Magnitudo (`mag`)** mengontrol **energi/guncangan**: semakin besar magnitudo, semakin besar intensitas ground motion. Banyak literatur teknik seismik menempatkan **Mw≈6** sebagai batas kejadian “besar” (rupture tidak lagi bisa dipandang titik).
* **Kedalaman (`depth`)** memodulasi guncangan di permukaan: untuk magnitudo yang sama, **gempa dangkal** menghasilkan guncangan permukaan **lebih kuat** daripada gempa yang lebih dalam. Karena itu **ambang magnitudo dinaikkan sedikit** untuk kelas kedalaman yang lebih dalam.

**Skema ambang operasional (siap pakai):**

* **Shallow (≤ 60 km)**

  * *kecil* : Mw < **4.5**
  * *sedang*: **4.5–<6.0**
  * *besar* : ≥ **6.0**
* **Intermediate (60–120 km)**

  * *kecil* : Mw < **4.7**
  * *sedang*: **4.7–<6.2**
  * *besar* : ≥ **6.2**
* **Deep (> 120 km)**

  * *kecil* : Mw < **4.9**
  * *sedang*: **4.9–<6.4**
  * *besar* : ≥ **6.4**

> **Intinya:** Klaster dilakukan di **ruang standar** agar fitur setara skala; **penamaan/label** dilakukan di **skala asli** memakai ambang di atas (atau konversi ambang ke z jika hanya z-score yang tersedia).
> **Longitude** & **latitude** dipakai untuk **posisi spasial** pada peta (visualisasi & struktur klaster), **bukan** penentu skala besar-kecil gempa.


### DBSCAN

```python
# ======================================
# 1. Import Library
# ======================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# ======================================
# Fungsi Klasifikasi Berdasarkan Threshold
# ======================================
def label_by_threshold_original(mw, depth_km):
    if depth_km <= 60:
        return 'kecil' if mw < 4.5 else ('sedang' if mw < 6.0 else 'besar')
    elif depth_km <= 120:
        return 'kecil' if mw < 4.7 else ('sedang' if mw < 6.2 else 'besar')
    else:
        return 'kecil' if mw < 4.9 else ('sedang' if mw < 6.4 else 'besar')

# ======================================
# 2. Baca Dataset
# ======================================
df = pd.read_csv('earthquake_subset.csv')
print("✅ Data berhasil dimuat:")
print(df.head())

# ======================================
# 3. Pilih Kolom yang Diperlukan
# ======================================
data = df[['longitude', 'latitude', 'mag', 'depth']].dropna()
print(f"Jumlah data setelah dibersihkan: {len(data)}")

# ======================================
# 4. Normalisasi Fitur (Magnitude & Depth)
# ======================================
scaler = StandardScaler()
X = scaler.fit_transform(data[['mag', 'depth']])

# ======================================
# 5. Terapkan DBSCAN
# ======================================
dbscan = DBSCAN(eps=0.7, min_samples=5)
data['cluster'] = dbscan.fit_predict(X)

print("\nJumlah data per cluster:")
print(data['cluster'].value_counts())

# ======================================
# 6. Tambahkan Label Berdasarkan Threshold Magnitude & Depth
# ======================================
data['cluster_label'] = data.apply(lambda row: label_by_threshold_original(row['mag'], row['depth']), axis=1)

# ======================================
# 7. Visualisasi Hasil Clustering
# ======================================
plt.figure(figsize=(10,7))

# Warna tetap untuk tiap kategori
colors = {'kecil':'blue', 'sedang':'orange', 'besar':'red'}

for label, color in colors.items():
    subset = data[data['cluster_label'] == label]
    plt.scatter(subset['longitude'], subset['latitude'],
                s=35, alpha=0.7, color=color, edgecolor='k', label=label)

plt.title('🌀 Klasifikasi Gempa di Indonesia\n(DBSCAN + Threshold Magnitude & Depth)', fontsize=13)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title="Tingkat Gempa")
plt.grid(True, linestyle='--', alpha=0.9)
plt.show()

# ======================================
# 8. Simpan Hasil Cluster (Opsional)
# ======================================
data.to_csv('hasil_cluster_gempa.csv', index=False)
print("\n💾 Hasil cluster disimpan sebagai hasil_cluster_gempa.csv")
```
