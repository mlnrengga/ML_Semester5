# Jelaskan kegunakan masing-masing library tersebut.
## 1. PyPREP
**Ringkasan**  
PyPREP adalah implementasi Python dari *PREP (Preprocessing Pipeline)* yang digunakan untuk preprocessing data EEG.  
Fungsinya terutama dalam mendeteksi dan memperbaiki channel bermasalah, melakukan rereferencing, serta filtering awal.  
Library ini bekerja erat dengan **MNE-Python** sehingga memudahkan peneliti EEG untuk membersihkan sinyal secara konsisten.  

**Kegunaan utama**  
- Deteksi channel noisy / bad channels secara otomatis.  
- Menstandarisasi pipeline preprocessing untuk dataset EEG berskala besar.  
- Integrasi dengan objek `mne.io.Raw` pada MNE-Python.  

---

## 2. SciPy
**Ringkasan**  
SciPy (*Scientific Python*) adalah library utama untuk komputasi ilmiah berbasis Python.  
Dibangun di atas NumPy, SciPy menyediakan modul untuk optimasi, integrasi, aljabar linear, statistik, pemrosesan sinyal, interpolasi, dan lainnya.  

**Kegunaan utama**  
- **Optimization**: pencarian minimum/maksimum fungsi.  
- **Signal processing**: filtering, transformasi, desain filter.  
- **Statistics**: distribusi probabilitas, uji statistik.  
- **Integration & ODE solver**: integrasi numerik, solusi persamaan diferensial.  
- **Linear Algebra**: operasi matriks, dekomposisi, sparse matrix.  

---

## 3. Weights & Biases (wandb)
**Ringkasan**  
Weights & Biases (wandb) adalah platform untuk *experiment tracking* dan manajemen proyek *Machine Learning*.  
wandb memudahkan visualisasi metrik training, pengelolaan hyperparameter, serta penyimpanan model/dataset.  

**Kegunaan utama**  
- Logging metrik training/validation secara real-time.  
- Visualisasi grafik, histogram, gambar, dan tabel hasil eksperimen.  
- Penyimpanan *artifacts* (model, dataset) untuk reproducibility.  
- Membandingkan eksperimen (*compare runs*) dan melakukan hyperparameter sweeps.  

---

## 4. pyECG
**Ringkasan**  
pyECG adalah library Python untuk bekerja dengan data **Electrocardiogram (ECG)**.  
Mendukung pembacaan/penulisan berbagai format dataset ECG serta menyediakan fungsi analisis seperti deteksi puncak R, interval QT, dan preprocessing sinyal ECG.  

**Kegunaan utama**  
- Membaca dan menulis format data ECG (misalnya WFDB, Holter).  
- Analisis interval (QT, RR), deteksi puncak R.  
- Pemrosesan sinyal ECG untuk penelitian kardiologi dan biometrik.

## 2. Berikan contoh tentang dampak energi dan lingkungan terhadap pemanfaatan kecerdasan buatan dan bagaimana cara mengatasinya (Anda dapat memberikan pendapat pribadi). Sertakan referensi yang digunakan.

## Dampak Energi dan Lingkungan
Pengembangan model AI berskala besar (misalnya Large Language Models) membutuhkan energi yang sangat besar.  
Studi **Strubell et al. (2019)** menunjukkan bahwa pelatihan satu model NLP dapat menghasilkan emisi karbon hingga setara dengan **5 kali emisi karbon rata-rata seorang warga Amerika selama hidupnya**.  
Selain emisi CO₂, pusat data juga membutuhkan konsumsi listrik tinggi dan air untuk sistem pendinginan, yang dapat membebani lingkungan sekitar.

## Cara Mengatasinya
Beberapa cara untuk mengurangi dampak tersebut antara lain:
- **Model compression**: menggunakan teknik distillation, pruning, atau quantization agar model lebih kecil dan hemat energi.  
- **Optimisasi hardware/software**: memakai chip khusus (TPU, GPU hemat daya) yang lebih efisien.  
- **Carbon-aware scheduling**: menjadwalkan pelatihan model di lokasi/waktu dengan pasokan energi terbarukan.  
- **Penggunaan energi terbarukan**: mendorong pusat data untuk memakai tenaga surya, angin, atau hidro.  

## Pendapat Pribadi
Menurut saya, solusi paling realistis untuk saat ini adalah **mengutamakan model yang lebih kecil dan efisien** dibanding selalu menggunakan model besar.  
Selain itu, perusahaan teknologi sebaiknya **transparan melaporkan jejak karbon** agar masyarakat tahu dampak lingkungan dari setiap inovasi AI.
