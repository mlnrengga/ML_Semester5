```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data Kelas TI-3G - Sheet1.csv')
```

```python
df.head(30)
```

#STANDARISASI DATA

```python
df['JENIS KELAMIN'] = df['JENIS KELAMIN'].replace(['Laki-laki', 'Laki-Laki', 'Laki - Laki', 'Laki Laki', 'Laki- Laki'], 'Laki-Laki')
df['JENIS KELAMIN'] = df['JENIS KELAMIN'].replace(['Perempuan'], 'Perempuan')

df['KOTA ASAL'] = df['KOTA ASAL'].str.replace('Kab. ', '', regex=False)
df['KOTA ASAL'] = df['KOTA ASAL'].str.replace('Kota ', '', regex=False)
df['KOTA ASAL'] = df['KOTA ASAL'].str.strip()

df['Asal Kelas (Kelas semester sebelumnya)'] = df['Asal Kelas (Kelas semester sebelumnya)'].str.replace('TI - ', '', regex=False)
df['Asal Kelas (Kelas semester sebelumnya)'] = df['Asal Kelas (Kelas semester sebelumnya)'].str.replace('TI-', '', regex=False)
df['Asal Kelas (Kelas semester sebelumnya)'] = df['Asal Kelas (Kelas semester sebelumnya)'].str.replace('TI ', '', regex=False)
df['Asal Kelas (Kelas semester sebelumnya)'] = df['Asal Kelas (Kelas semester sebelumnya)'].str.strip()

df['Target IPS (target realistis IPS semester ini)'] = df['Target IPS (target realistis IPS semester ini)'].astype(str).str.replace(',', '.', regex=False)
df['Target IPS (target realistis IPS semester ini)'] = pd.to_numeric(df['Target IPS (target realistis IPS semester ini)'])
```

```python
df.head(30)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['JENIS KELAMIN'] = le.fit_transform(df['JENIS KELAMIN'])
```
